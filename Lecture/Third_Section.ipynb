## Numpy ndarray

* This is the [key datatype](https://numpy.org/doc/stable/reference/generated/numpy.ndarray.html) of numpy.
* Similar to a list but expects all data to be numeric and the same type.
* You can create an ndarray from lists 
* Lots of features to [ndarray](https://numpy.org/devdocs/user/basics.creation.html)

```python
vector = np.array([1,2,3,4,5], dtype=np.int8)
print(f"Vector 1D = {vector} {vector.shape}")

vector2d = np.array([[1,2,3],[4,5,6]], dtype=np.complex128)
print(f"Vector 2D = {vector2d}")

vector3d = np.ones((2, 3, 2))  # or np.zeros
print(f"Vector 3D = {vector3d}")
```

```python
A = np.array([[1, 1],
              [0, 1]])
B = np.array([[2, 0],
              [3, 4]])
print(f"Add \n{A + B}\n")
print(f"Element Mult \n{A * B}\n")     # elementwise product
print(f"Dot Product \n{A @ B}\n")     # matrix product
print(f"Dot Product Method \n{A.dot(B)}\n")
```

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
%matplotlib inline
matplotlib.__version__
```

The %matplotlib inline command is a special function for making matplotlib work correctly with Jupyter.  Without it the plots get popped up in a separate window instead of being in the page.

- [Plotting](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)
- plot and pair of x and y values.
- add to the [legend](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html) in the order added to the plot
- Can set [x](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlabel.html) and [y](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylabel.html) axis labels
- Can set a [title](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.title.html)
- [Grid](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.grid.html) can be enabled or disabled
- Need to show the plot to get it to display

```python
x_vals = np.linspace(-2*np.pi, 2*np.pi,100)
sin_wave = np.sin(x_vals)
cos_wave = np.cos(x_vals)

fig, ax = plt.subplots()

ax.plot(x_vals, sin_wave, 'ro')
ax.plot(x_vals, cos_wave,)
ax.legend(["sine", "cosine"])
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Plot Title")
plt.grid(True)
plt.show()
```

You can also visualize [multiple plots](https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html) in the same image.

```python
fig, axs = plt.subplots(2, sharex=True, sharey=True)
fig.suptitle('Sharing both axes')
axs[0].plot(x_vals, sin_wave, 'g')
axs[1].plot(x_vals, cos_wave, 'r')
plt.show()
```

## Pandas DataFrame

* [Documentation](https://pandas.pydata.org/docs/reference/frame.html)
* [Start Up](https://pandas.pydata.org/docs/user_guide/dsintro.html#dataframe)
* This is the key data structure of pandas
* a 2-dimensional labeled data structure with columns of potentially different types. 
* You can think of it like a spreadsheet or SQL table.
* Can quickly load a csv file, [read_csv](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html)

```python
stock_data_file = "all_stocks_5yr.csv"
stock_data_frame = pd.read_csv(stock_data_file)
```

The [info](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.info.html#pandas.DataFrame.info) method prints information about a DataFrame including the index dtype and columns, non-null values and memory usage.

```python
stock_data_frame.info()
```

The [columns](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.columns.html) field holds a list of the labels for the columns

```python
stock_data_frame.columns
```

The [head](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.head.html) method This function returns the first n rows for the object based on position. It is useful for quickly testing if your object has the right type of data in it.

```python
stock_data_frame.head(10)
```

The [query](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.query.html) method returns the specified columns

```python
def get_stock_by_symbol(name=None):
    if name is None:
        return None
    # This is risky code!  What if the stock symbol passed in does not exist?
    # Silently return an empty frame
    return stock_data_frame.query(f"Name == '{name}'")
```

```python
stock = get_stock_by_symbol(name="GOOG")
print(stock)
```

```python
np.mean(stock)
```

```python
stock["close"].plot(legend=True,figsize=(10,4))
```

- Matplotlib has special [date](https://matplotlib.org/stable/api/dates_api.html) formatting handlers
- There are tools for formatting the date with [autofmt_xdate](https://matplotlib.org/stable/api/figure_api.html#matplotlib.figure.Figure.autofmt_xdate)
- Pandas can convert to the correct date time formate with [to_datetime](https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html)
- [gca](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.gca.html) is Get Current Axis 
- [gcf](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.gcf.html) is Get Current Figure

```python
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y/%m/%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=90))
plt.gcf().autofmt_xdate()
plt.plot(pd.to_datetime(stock["date"].values), stock["high"].values)
plt.plot(pd.to_datetime(stock["date"].values), stock["low"].values)
plt.legend(["high", "low"])
plt.xlabel("Dates")
plt.ylabel("Price ($)")
plt.title(f"Stock High and Low of {stock['Name'].values[0]}")
plt.grid(True)
plt.show()
```
